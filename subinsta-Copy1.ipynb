````python
template_path="pst_template.jpg"
base_folder="Big_insta_posts"
output_folder="insta_generated_posts"
os.makedirs(output_folder,exist_ok=True)
````

````python
font_path="C:/Windows/Fonts/arialbd.ttf"
font=ImageFont.truetype(font_path,size=50)
header_font=ImageFont.truetype(font_path,size=100)
````

````python


list_header = ["updated_1.jpg", "updated_2.jpg", "updated_3.jpg"]
footer_image = "end_updated.jpg"

line_spacing = 10
question_y = 80

# -------- UTILITIES --------
def wrap_text(text, font, max_width):
    words = text.split()
    lines = []
    current_line = ""
    for word in words:
        test_line = f"{current_line} {word}".strip()
        if font.getlength(test_line) <= max_width:
            current_line = test_line
        else:
            lines.append(current_line)
            current_line = word
    if current_line:
        lines.append(current_line)
    return lines

def draw_centered_text(image, text, font, max_width_ratio=0.9):
    draw = ImageDraw.Draw(image)
    w, h = image.size

    # Get the font file path and starting size
    font_path = font.path  # works because you used truetype()
    current_font_size = font.size

    max_width = w * max_width_ratio

    # Decrease font size until text fits
    while True:
        temp_font = ImageFont.truetype(font_path, current_font_size)
        text_width = temp_font.getlength(text)
        if text_width <= max_width or current_font_size <= 10:
            break
        current_font_size -= 2  # step down

    # Calculate position for final text
    text_height = temp_font.getbbox(text)[3] - temp_font.getbbox(text)[1]
    position = ((w - text_width) // 2, (h - text_height) // 2)
    draw.text(position, text, font=temp_font, fill="black")

    return image

# -------- MAIN LOOP --------
os.makedirs(output_folder, exist_ok=True)

for topic in os.listdir(base_folder):
    topic_path = os.path.join(base_folder, topic)
    if os.path.isdir(topic_path):
        topic_output = os.path.join(output_folder, topic)
        os.makedirs(topic_output, exist_ok=True)

        # Header image
        random_header = random.choice(list_header)
        header_img = Image.open(random_header).convert("RGB")
        header_img = draw_centered_text(header_img, topic, header_font)
        header_img.save(os.path.join(topic_output, "000_header.jpg"))

        index = 1
        for file_name in sorted(os.listdir(topic_path)):
            if file_name.endswith(".txt"):
                with open(os.path.join(topic_path, file_name), "r", encoding="utf-8") as f:
                    lines = f.readlines()
                    question = lines[0].strip()
                    answer = " ".join(line.strip() for line in lines[1:])

                img = Image.open(template_path).convert("RGB")
                img_width, img_height = img.size
                padding = 60
                text_area_width = img_width - 2 * padding
                draw = ImageDraw.Draw(img)

                # Draw Question
                wrapped_q = wrap_text(question, font, text_area_width)
                y_q = question_y
                for line in wrapped_q:
                    draw.text((padding, y_q), line, font=font, fill="black")
                    line_height = font.getbbox(line)[3] - font.getbbox(line)[1]
                    y_q += line_height + line_spacing

                # Draw Answer with bullet points
                answer_sentences = [sent.strip() for sent in answer.split('. ') if sent.strip()]
                wrapped_a_lines = []
                for sent in answer_sentences:
                    bullet_line = "\u2022 " + sent
                    wrapped_lines = wrap_text(bullet_line, font, text_area_width)
                    wrapped_a_lines.extend(wrapped_lines)
                    wrapped_a_lines.append("")

                y_a = y_q + 230
                for line in wrapped_a_lines:
                    draw.text((padding, y_a), line, font=font, fill="black")
                    line_height = font.getbbox("A")[3] - font.getbbox("A")[1]
                    y_a += line_height + line_spacing

                # Save post
                output_file = os.path.join(topic_output, f"{str(index).zfill(3)}_{file_name.replace('.txt', '')}.jpg")
                img.save(output_file)
                index += 1

        # Footer image
        footer_img = Image.open(footer_image).convert("RGB")
        footer_img.save(os.path.join(topic_output, f"{str(index).zfill(3)}_footer.jpg"))

print("✅ All images generated with large topic headers and footers. Check 'inata_generated_posts/'.")
````
